```python
# Question 3
# Load the original color image
color_image = cv2.imread('images/lena.jfif')

# Convert the color image to grayscale
gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)

# Resize grayscale image to match color image's dimensions
gray_image = cv2.resize(gray_image, (color_image.shape[1], color_image.shape[0]))

# Concatenate the images horizontally
combined_images = cv2.hconcat([color_image, cv2.cvtColor(gray_image, cv2.COLOR_GRAY2BGR)])

# Display the combined image using OpenCV
cv2.imshow('Color and Grayscale Images', combined_images)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the original image
image_path = 'images/dog.jfif'
original_image = cv2.imread(image_path)

# Define the scaling factor
scaling_factor = 2  # You can adjust this value

# Calculate the new dimensions for the enlarged image
new_width = int(original_image.shape[1] * scaling_factor)
new_height = int(original_image.shape[0] * scaling_factor)

# Perform linear interpolation
linear_interpolation = cv2.resize(original_image, (new_width, new_height), interpolation=cv2.INTER_LINEAR)

# Perform cubic interpolation
cubic_interpolation = cv2.resize(original_image, (new_width, new_height), interpolation=cv2.INTER_CUBIC)

# Perform nearest neighbor interpolation
nearest_neighbor_interpolation = cv2.resize(original_image, (new_width, new_height), interpolation=cv2.INTER_NEAREST)

# Display the original and enlarged images using matplotlib
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(linear_interpolation, cv2.COLOR_BGR2RGB))
plt.title('Linear Interpolation')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(cubic_interpolation, cv2.COLOR_BGR2RGB))
plt.title('Cubic Interpolation')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(nearest_neighbor_interpolation, cv2.COLOR_BGR2RGB))
plt.title('Nearest Neighbor Interpolation')

plt.tight_layout()
plt.show()
```
